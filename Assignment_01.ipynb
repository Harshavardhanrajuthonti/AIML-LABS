1.1 ) Lists

```python
import numpy as np
import pandas as pd

my_list = [1, 2, 3, 4, 5]

numpy_array = np.array(my_list)
pandas_series = pd.Series(my_list)

print("Type of NumPy array:", type(numpy_array))
print("Type of Pandas series:", type(pandas_series))
```

1.6) Implement np.sin(), log(), log2(), np.exp())

3.3) Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({'People': people, 'Age': age, 'Height': height, 'Weight': weight})

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(healthcare_data['Age'], healthcare_data['Height'], color='blue', label='Height')
plt.scatter(healthcare_data['Age'], healthcare_data['Weight'], color='red', label='Weight')
plt.title('Scatter Plot of Age vs Height and Weight')
plt.xlabel('Age')
plt.ylabel('Measurement')
plt.legend()
plt.grid(True)
plt.show()
```

2.4) Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({'People': people, 'Age': age, 'Height': height, 'Weight': weight})

plt.figure(figsize=(8, 6))
plt.bar(healthcare_data['People'], healthcare_data['Height'], color='blue', label='Height')
plt.bar(healthcare_data['People'], healthcare_data['Weight'], color='red', label='Weight')
plt.title('Bar Chart of Height and Weight by Person')
plt.xlabel('People')
plt.ylabel('Measurement')
plt.legend()
plt.grid(axis='y')
plt.show()
```

2.5) Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({'People': people, 'Age': age, 'Height': height, 'Weight': weight})

plt.figure(figsize=(8, 6))
plt.hist(healthcare_data['Age'], bins=5, color='blue', alpha=0.7, label='Age')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y')
plt.show()
```

2.6) Provide Graph Title, labels for X, Y axis with proper justification and explaination of
the graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

healthcare_data = pd.DataFrame({'People': people, 'Age': age, 'Height': height, 'Weight': weight})

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(healthcare_data['Age'], bins=5, color='blue', alpha=0.7)
plt.title('Distribution of Age in Healthcare Dataset', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--')

plt.text(30, 1.5, 'Explanation: \n\nThis histogram represents the distribution of ages \nacross individuals in the healthcare dataset. \nEach bar represents a range of ages, and the \nheight of the bar indicates the frequency of \nindividuals within that age range.', fontsize=12)

plt.show()

```
